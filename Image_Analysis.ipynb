```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mimg
import numpy as np
```

```python
from skimage import color
from skimage import feature
import glob#For bulk image importing
```

```python
# Set the default paramenter
plt.rc['image.cmap'] = 'gray' # Display the images in grayscale
plt.rc['figure.figsize'] = 10, 10
```

```python
# Importing the images from the folder of images
imgpaths = glob.glob("imagelist/python_logo.png")+glob.glob("imagelist/Anaconda_logo.png")+glob.glob("imagelist/Sunrise Image.jpg")
```

```python
imgageset = [mimg.imread(x) for x in imgpaths]
```

```python
#To display the thumb nails of the image along with the co ordinates plots when they are loaded by showing the coordinates in more detail
plt.figure()
for i,img in enumerate(imageset):#enum is used as we are making a choice and iterating over through a list of images
    plt.subplot(1, len(imageset), i+1)
    plt.imshow(img, cmap = 'gray')
```

```python
#To show each of the image separately when they are loaded in a larger way from the list of images
for i,img in enumerate(imageset):
    plt.figure()
    plt.imshow(img)
```

```python
from skimage import img_as_float
```

```python
for i,img in enumerate(imageset):
    img_bw = img_as_float(color.rgb2grey(img))
    plt.figure()
    plt.subplot(1, 2, 1) # One row, two columns ofimage 1 mostly of the normal scale based on the index
    plt.imshow(img)
    plt.subplot(1, 2, 2)  # One row, two columns of image 2 mostly of the grayscale image based on the index
    plt.imshow(img_bw)
```

```python
#To print the dimensions, shape and size of the images
for i,img in enumerate(imageset):
    print(img.ndim)
    print(img.size)
    print(img.shape)
    print(np.amax(img))#For returning the maximum of the array or in the index
```

```python
#Check the shape size and dimensions of the grayscale or the black and white images
for i,img in enumerate(img_bw):
    print(img.ndim)
    print(img.size)
    print(img.shape)
    print(np.amax(img))
```

### Note that the Black and White image dimension is reduced to just 1 from the color images whose dimensions is 3

```python
# Find horizontal edges using a simple shifting method
def find_horizontal_edges(img):
    imgbw = img_as_float(color.rgb2grey(img))
    return np.abs(imgbw[:, 1:] - imgbw[:, :-1])

# Find vertical edges using a simple shifting method
def find_vertical_edges(img):
    imgbw = img_as_float(color.rgb2grey(img))
    return np.abs(imgbw[1:, :] - imgbw[:-1, :])
```
